```python
import pandas as pd
```

```python
dataset = pd.read_csv("SPAM text message 20170820 - Data.csv")
```

```python
dataset
```

```python
dataset.shape
```

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(dataset)):
    review = re.sub('[^a-zA-z0-9]'," ",dataset['Message'][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## Creating bag of words model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,binary=True)
x = cv.fit_transform(corpus).toarray()
```

```python
x
```

```python
y = pd.get_dummies(dataset['Category'],dtype=int)
```

```python
y = y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB().fit(x_train,y_train)
```

```python
ypred = spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,f1_score
```

```python
accuracy_score(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```

```python
f1_score(y_test,ypred)
```

## word2vec from scratch

```python
corpus[0]
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
words = []
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
words[0]
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(words,window=5,min_count=2)
```

```python
model.wv.index_to_key
```

```python
model.corpus_count
```

```python
model.epochs
```

```python
model.wv.similar_by_word('happy')
```

```python
import numpy as np
```

```python
def avg_word2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key],axis=0)
```
